```python
df6["char_length"]=df6["lines"].apply(lambda x:len(str(x)))
```

```python
final_df=pd.concat([df,df2,df3,df4,df5,df6])
```

```python
final_df.reset_index(inplace=True)
```

```python
final_=final_df.dropna()
```

```python
final_.reset_index(inplace=True)
```

```python
final_["label"].value_counts()
```

```python
final_.head()
```

```python
used=['char_length', 'no_of_words',
       'caps', 'space', 'next_char', 'next_words', 'next_next_char',
       'next_next_words', 'next_caps', 'prev_char', 'prev_words',
       'prev_prev_char', 'prev_prev_words']
```

```python
 X_train, X_test, y_train, y_test = train_test_split(final_[used], final_['label'], test_size=0.2, random_state=42,stratify=final_['label'])
```

```python
y_train.value_counts()
```

```python

xgb=XGBClassifier()

for mod in [xgb]:
    mod.fit(X_train,y_train)
    pred=mod.predict(X_test)
    print(classification_report(y_test,pred))
    
# 71 percent F1 Score on heading.
```

```python
title=final_[final_["label"]==1].sample(n=600,random_state=42,replace=True)
section=final_[final_["label"]==0]
data =pd.concat([title,section])
```

```python
xgb=XGBClassifier()
xgb.fit(data[used],data['label'])
```
